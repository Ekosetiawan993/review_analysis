### Use the stemmed reviews with minimal stopwords removed away

### Context
#### I decided to convert rating 1 and 2 as "Negative" sentiment, 3 as "Neutral", 4 and 5 as "Positive"

```python
data["review_sentiment"] = data.rating.map({
    1 : "Negative",
    2 : "Negative",
    3 : "Neutral",
    4 : "Positive",
    5 : "Positive"
})
```

```python
data_v3["review_sentiment"] = data_v3.rating.map({
    1 : "Negative",
    2 : "Negative",
    3 : "Neutral",
    4 : "Positive",
    5 : "Positive"
})
```

### Split data to train and test

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(
    data.stemmed_review, 
    data.review_sentiment,
    test_size=0.2,
    stratify=data.review_sentiment,
    random_state=50
)
```

```python
test_x.shape
```

```python
test_y.value_counts()
```

```python
#import TfidfVectorizer from sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf_idf = TfidfVectorizer()
```

```python
transformed = tf_idf.fit_transform(train_x)
```

```python
tf_idf.vocabulary_
```

```python
#let's print the idf of each word:

all_feature_names = tf_idf.get_feature_names()

for word in all_feature_names:
    
    #let's get the index in the vocabulary
    indx = tf_idf.vocabulary_.get(word)
    
    #get the score
    idf_score = tf_idf.idf_[indx]
    
    print(f"{word} : {idf_score}")
```

### Use TF-IDF and MultinomialNB

```python
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
```

```python
multi_nb_pipe = Pipeline(steps=[
    ('tf-idf', TfidfVectorizer()),
    ('multi_nb', MultinomialNB()),
])
```

```python
multi_nb_pipe.fit(train_x, train_y)
```

```python
multi_nb_prediction = multi_nb_pipe.predict(test_x)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(test_y, multi_nb_prediction))
```

### Using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Create pipeline
random_forest_pipe = Pipeline(steps=[
    ('tf-idf', TfidfVectorizer()),
    ('random forest', RandomForestClassifier(random_state=50)),
])

# Training
random_forest_pipe.fit(train_x, train_y)

# Make prediction
random_forest_prediction = random_forest_pipe.predict(test_x)

print(classification_report(test_y, random_forest_prediction))
```

### Using Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
# Create pipeline
gradient_boosting_pipe = Pipeline(steps=[
    ('tf-idf', TfidfVectorizer()),
    ('random forest', GradientBoostingClassifier(random_state=50)),
])

# Training
gradient_boosting_pipe.fit(train_x, train_y)

# Make prediction
gradient_boosting_prediction = gradient_boosting_pipe.predict(test_x)

print(classification_report(test_y, gradient_boosting_prediction))
```

### Using Multi Layer Perceptron

```python
from sklearn.neural_network import MLPClassifier
```

```python
# Create pipeline
mlp_pipe = Pipeline(steps=[
    ('tf-idf', TfidfVectorizer()),
    ('MLP', MLPClassifier(random_state=50)),
])

# Training
mlp_pipe.fit(train_x, train_y)

# Make prediction
mlp_prediction = mlp_pipe.predict(test_x)

print(classification_report(test_y, mlp_prediction))
```
