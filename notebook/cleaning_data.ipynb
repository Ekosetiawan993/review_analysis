### Looks like we have some problems with new line (\n) and emoticon

```python
import re
```

```python
data.iloc[37, :]
```

```python
data_2 = data.copy()
```

### Delete emoticon

```python
data_2['review_2'] = data_2.review.str.replace('[^A-Za-z0-9\s,.-]', '', flags=re.UNICODE, regex=True)
```

```python
data_2.loc[35:40, :]
```

```python
data_2.info()
```

**Delete new line \n**

```python
data.iloc[40,:]
```

```python
data.iloc[40,:].str.replace('\n', ' ')
```

```python
data.iloc[40,0].replace('..', '.')
```

```python
data_2['review_3'] = data_2.review_2.str.replace('\n', ' ')
```

**Replace double period .. with single period**

```python
data_2['review_4'] = data_2.review_3.str.replace('..', '.', regex=False)
```

```python
data_2.iloc[40,:]
```

```python
data_2.loc[35:40, :]
```

#### review_4 is the cleaner review, let's put it to our main data
